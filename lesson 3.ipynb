```python
from math import factorial, sqrt
import numpy as np
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

## –£—Ä–æ–∫ 3. –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞. –ö–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –ø–æ–ø—É–ª—è—Ü–∏–∏. –ì—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö

### 1. –î–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –∏–∑ –≤—ã–±–æ—Ä–∫–∏ –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. –ü–æ—Å—á–∏—Ç–∞—Ç—å (–∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –Ω–∞–ø–æ–¥–æ–±–∏–µ std, var, mean) —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ, —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ, —Å–º–µ—â–µ–Ω–Ω—É—é –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.

```python
sample = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
n = sample.shape[0]
```

```python
M = sample.sum() / n

root_sum = sum(map(lambda x: pow(x - M, 2), sample))

square_S_1 = root_sum / n
square_S_2 = root_sum / (n - 1)

S = sqrt(square_S_2)
```

```python
print(
f'''
–°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ M = {M}
–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏ S = {S}
–°–º–µ—â–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è S—Å–º^2 = {square_S_1}
–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è S–Ω^2 = {square_S_2}
''')
```

### 2. –í –ø–µ—Ä–≤–æ–º —è—â–∏–∫–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è 8 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 - –±–µ–ª—ã–µ. –í–æ –≤—Ç–æ—Ä–æ–º —è—â–∏–∫–µ - 12 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 –±–µ–ª—ã—Ö. –ò–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ –≤—ã—Ç–∞—Å–∫–∏–≤–∞—é—Ç —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –¥–≤–∞ –º—è—á–∞, –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ - 4. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ 3 –º—è—á–∞ –±–µ–ª—ã–µ?

### –†–µ—à–µ–Ω–∏–µ:

–ü—É—Å—Ç—å  

$ P_1 - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ç–∞—â–∏—Ç—å –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ 2 –±–µ–ª—ã—Ö –º—è—á–∞ –∏ –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ 1 –±–µ–ª—ã–π –∏ 3 –Ω–µ –±–µ–ª—ã—Ö   
$ P_2 - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ç–∞—â–∏—Ç—å –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ 1 –±–µ–ª—ã–π –º—è—á –∏ 1 –Ω–µ –±–µ–ª—ã–π –∏ –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ 2 –±–µ–ª—ã—Ö –∏ 2 –Ω–µ –±–µ–ª—ã—Ö  
$ P_3 - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ç–∞—â–∏—Ç—å –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ 0 –±–µ–ª—ã—Ö –º—è—á–µ–π –∏ –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ 3 –±–µ–ª—ã—Ö –º—è—á–∞ –∏ 1 –Ω–µ –±–µ–ª—ã–π  

```python
P_1 = combinations(5, 2) / combinations(8, 2) * combinations(5, 1) * combinations(7, 3) / combinations(12, 4)
P_2 = combinations(5, 1) * combinations(3, 1) / combinations(8, 2) * combinations(5, 2) * combinations(7, 2) / combinations(12, 4)
P_3 = combinations(3, 2) / combinations(8, 2) * combinations(5, 3) * combinations(7, 1) / combinations(12, 4)

P = P_1 + P_2 + P_3
```

```python
print(f'P_1 = {P_1}\nP_2 = {P_2}\nP_3 = {P_3}\nP = {P}')
```

### 3. –ù–∞ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ –±–∏–∞—Ç–ª–æ–Ω—É –æ–¥–∏–Ω –∏–∑ —Ç—Ä–µ—Ö —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ —Å—Ç—Ä–µ–ª—è–µ—Ç –∏ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –º–∏—à–µ–Ω—å. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞ —Ä–∞–≤–Ω–∞ 0.9, –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ ‚Äî 0.8, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–≥–æ ‚Äî 0.6. –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω: a). –ø–µ—Ä–≤—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º –±). –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º –≤). —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º.

### –†–µ—à–µ–Ω–∏–µ:

–ü—É—Å—Ç—å  

$ P(B_1) - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω –ø–µ—Ä–≤—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º  
$ P(B_2) - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º  
$ P(B_3) - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º  

—Ç–æ–≥–¥–∞ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ–æ—Ä–º—É–ª–æ–π –ë–∞–π–µ—Å–∞  

$ P(B_1) = \frac{\frac{0.9}{3}}{\frac{0.9}{3}\cdot\frac{0.8}{3}\cdot\frac{0.6}{3}} $

```python
Pb_1 = (0.9 / 3) / ((0.9 + 0.8 + 0.6) / 3)
Pb_2 = (0.8 / 3) / ((0.9 + 0.8 + 0.6) / 3)
Pb_3 = (0.6 / 3) / ((0.9 + 0.8 + 0.6) / 3)
```

```python
print(
f'''
–∞) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω –ø–µ—Ä–≤—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {Pb_1}
–±) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {Pb_2}
–≤) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {Pb_3}

–ü—Ä–æ–≤–µ—Ä–∫–∞: –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª —Å–æ–≤–µ—Ä—à–∏–ª –æ–¥–∏–Ω –∏–∑ —Ç—Ä–µ—Ö —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_A + P_B + P_C}
''')
```

### 4. –í —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç—ã A –∏ B –ø–æ—Å—Ç—É–ø–∏–ª–æ —Ä–∞–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∞ –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç C —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ—Å—Ç—É–ø–∏–ª–æ —Å—Ç–æ–ª—å–∫–æ –∂–µ, —Å–∫–æ–ª—å–∫–æ –Ω–∞ A –∏ B –≤–º–µ—Å—Ç–µ. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ A —Å–¥–∞—Å—Ç –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é, —Ä–∞–≤–Ω–∞ 0.8. –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ B —ç—Ç–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 0.7, –∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ C - 0.9. –°—Ç—É–¥–µ–Ω—Ç —Å–¥–∞–ª –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–Ω —É—á–∏—Ç—Å—è: a). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A –±). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B –≤). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C?

–†–µ—à–µ–Ω–∏–µ:  
–ü—É—Å—Ç—å  

$ ùëÉ(A) ‚àí $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A  
$ ùëÉ(ùêµ) ‚àí $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B  
$ ùëÉ(C) ‚àí $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C  

$ p_1 - $ –∞–ø—Ä–∏–æ—Ä–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A  
$ p_2 - $ –∞–ø—Ä–∏–æ—Ä–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B  
$ p_3 - $ –∞–ø—Ä–∏–æ—Ä–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C  

—Ç–æ–≥–¥–∞ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ–æ—Ä–º—É–ª–æ–π –ë–∞–π–µ—Å–∞  

```python
p_1 = 0.25
p_2 = 0.25
p_3 = 0.5

P_A = (0.8 * p_1) / (0.8 * p_1 + 0.7 * p_2 + 0.9 * p_3)
P_B = (0.7 * p_2) / (0.8 * p_1 + 0.7 * p_2 + 0.9 * p_3)
P_C = (0.9 * p_3) / (0.8 * p_1 + 0.7 * p_2 + 0.9 * p_3)
```

```python
print(
f'''
–∞) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_A}
–±) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_B}
–≤) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_C}

–ü—Ä–æ–≤–µ—Ä–∫–∞: –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ –æ–¥–Ω–æ–º –∏–∑ —Ç—Ä–µ—Ö —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_A + P_B + P_C}
''')
```

### 5. –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Ç—Ä–µ—Ö –¥–µ—Ç–∞–ª–µ–π. –î–ª—è –ø–µ—Ä–≤–æ–π –¥–µ—Ç–∞–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–π—Ç–∏ –∏–∑ —Å—Ç—Ä–æ—è –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü —Ä–∞–≤–Ω–∞ 0.1, –¥–ª—è –≤—Ç–æ—Ä–æ–π - 0.2, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–π - 0.25. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –≤—ã–π–¥—É—Ç –∏–∑ —Å—Ç—Ä–æ—è: –∞). –≤—Å–µ –¥–µ—Ç–∞–ª–∏ –±). —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏ –≤). —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å –≥). –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π?

### –†–µ—à–µ–Ω–∏–µ:  
–ü—É—Å—Ç—å  

$ P_1 - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è –≤—Å–µ—Ö –¥–µ—Ç–∞–ª–µ–π  
$ P_2 - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π  
$ P_3 - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–π –¥–µ—Ç–∞–ª–∏  
$ P_4 - $ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è –æ–¥–Ω–æ–π –∏–ª–∏ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π  

```python
P_1 = 0.1 * 0.2 * 0.25
P_2 = 0.1 * 0.2 + 0.1 * 0.25 + 0.2 * 0.25
P_3 = 1 - (1 - 0.1) * (1 - 0.2) * (1 - 0.25)
P_4 = P_3 - P_1
```

```python
print(
f'''
–∞) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è –≤—Å–µ—Ö –¥–µ—Ç–∞–ª–µ–π —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_1}
–±) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_2}
–≤) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–π –¥–µ—Ç–∞–ª–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_3}
–≥) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {P_4}
''')
```
